```python
import pandas as pd
df = pd.read_csv('data.csv')
print(df)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].mean()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].median()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].mode()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].max()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].min()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].max()-df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].min()
print(df2)
```

```python
df2 = (df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].max()+df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].min())/2
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].var()
print(df2)
```

```python
df2 = df[["smoothness_mean","compactness_mean","concavity_mean","texture_mean"]].std()
print(df2)
```

```python
data = pd.read_csv('data.csv')
data.dtypes
```

```python
# calculate a 5-number summary
from numpy import percentile
from numpy.random import rand
# generate data sample
df = rand(1000)
# calculate quartiles
quartiles = percentile(df, [25, 50, 75])
# calculate min/max
data_min, data_max = data.min(), data.max()
# print 5-number summary
print('Min: %.3f' % data_min)
print('Q1: %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % data_max)
```
